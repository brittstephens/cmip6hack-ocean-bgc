Plot `o2`, `o2sat`, and Apparent Oxygen Utilization (although at the surface, it shows disequilibrium rather than biogenic consumption)

```python
fig, (ax1,ax2,ax3) = plt.subplots(figsize=(15,4),ncols=3,nrows=1)
(model.isel(time=slice(-13,-1)).mean('time').isel(lev=0).o2*1e3).plot(ax=ax1,levels=np.arange(200,420,20),extend="both")

# Looks like O2sat is 1000x too small! Maybe they doubley applied the mmol->mol conversion?
(model.isel(time=slice(-13,-1)).mean('time').isel(lev=0).o2sat*1e6).plot(ax=ax2,levels=np.arange(200,420,20),extend="both")
aou=(model.o2sat*1000-model.o2)
(aou.isel(time=slice(-13,-1)).mean('time').isel(lev=0)*1e3).plot(ax=ax3,levels=np.arange(-100,120,20),extend="both")

ax1.set_title("Oxygen Concentration [mmol/m3]")
ax2.set_title("Oxygen Saturation [mmol/m3]")
ax3.set_title("Apparent Oxygen Utilization [mmol/m3]")
plt.show()
```

Derive Preformed and Regenerated carbon (assuming Redfield ratios), and plot Atlantic and Pacific sections

```python
R_co=117/170

cpre=(model.dissic-aou*R_co)*1e3
creg=aou*R_co*1e3

fig, (axes) = plt.subplots(figsize=(10,10),ncols=2,nrows=3)
(model.isel(time=slice(-13,-1)).mean('time').isel(i=255).dissic*1e3).plot(ax=axes[0,0],levels=np.arange(2000,2420,20),cmap='viridis',extend="both")
(model.isel(time=slice(-13,-1)).mean('time').isel(i=125).dissic*1e3).plot(ax=axes[0,1],levels=np.arange(2000,2420,20),cmap='viridis',extend="both")

(cpre.isel(time=slice(-13,-1)).mean('time').isel(i=255)).plot(ax=axes[1,0],levels=np.arange(2000,2420,20),cmap='viridis',extend="both")
(cpre.isel(time=slice(-13,-1)).mean('time').isel(i=125)).plot(ax=axes[1,1],levels=np.arange(2000,2420,20),cmap='viridis',extend="both")

(creg.isel(time=slice(-13,-1)).mean('time').isel(i=255)).plot(ax=axes[2,0],levels=np.arange(0,320,20),cmap='viridis',extend="both")
(creg.isel(time=slice(-13,-1)).mean('time').isel(i=125)).plot(ax=axes[2,1],levels=np.arange(0,320,20),cmap='viridis',extend="both")

for jj in range(0,2):
    for ii in range(0,3):
        if jj==0:
            axes[ii,jj].set_xlim(330,0)
        axes[ii,jj].set_ylim(6000,0)
        axes[ii,jj].set_xlabel("")
        
axes[0,0].set_title("Atlantic DIC Concentration [mol/m3]")
axes[0,1].set_title("Pacific DIC Concentration [mmol/m3]")
axes[1,0].set_title("Atlantic Cpre [mmol/m3]")
axes[1,1].set_title("Pacific Cpre [mmol/m3]")
axes[2,0].set_title("Atlantic Creg [mmol/m3]")
axes[2,1].set_title("Pacific Creg [mmol/m3]")
plt.show()
```

# What about models that dont have o2sat? Can calculate using the GSW toolbox

```python
import gsw
```

```python
#Calculate Absolute Salinity
models['gsw_as']=xr.apply_ufunc(gsw.SA_from_SP,
                                models.so.isel(time=slice(-13,-1)).mean('time'), 
                                models.lev, 
                                models.longitude, 
                                models.latitude, 
                                dask='parallelized',
                                output_dtypes=[models.so.dtype])

fig, (ax1,ax2,ax3) = plt.subplots(figsize=(15,4),ncols=3,nrows=1)
(models.isel(time=slice(-13,-1)).mean('time').isel(lev=0).so).plot(ax=ax1,levels=np.arange(30,40.5,0.5),extend="both")
ax1.set_title("Practical Salinity [psu] from model")
(models.isel(lev=0).gsw_as).plot(ax=ax2,levels=np.arange(30,40.5,0.5),extend="both")
ax2.set_title("Absolute Salinity [g/kg] from GSW")
sanom=(models.so-models.gsw_as)
(sanom.isel(time=slice(-13,-1)).mean('time').isel(lev=0)).plot(ax=ax3,levels=np.arange(-0.2,0.220,0.020),extend="both")
ax3.set_title("Practical Salinity-Absolute Salinity from GSW")
plt.show()
```

```python
#Calculate Conservative Temperature

modelt['gsw_ct']=xr.apply_ufunc(gsw.CT_from_t, models.gsw_as, 
                                modelt.thetao.isel(time=slice(-13,-1)).mean('time'),
                                modelt.lev, 
                                dask='parallelized',
                                output_dtypes=[modelt.thetao.dtype])

fig, (ax1,ax2,ax3) = plt.subplots(figsize=(15,4),ncols=3,nrows=1)
(modelt.isel(time=slice(-13,-1)).mean('time').isel(lev=0).thetao).plot(ax=ax1,levels=np.arange(-2,38,2),extend="both")
ax1.set_title("Potential Temperature [°C] from model")
(modelt.isel(lev=0).gsw_ct).plot(ax=ax2,levels=np.arange(-2,38,2),extend="both")
ax2.set_title("Conservative Temperature [°C] from GSW")
tanom=(modelt.thetao-modelt.gsw_ct)
(tanom.isel(time=slice(-13,-1)).mean('time').isel(lev=0)).plot(ax=ax3,levels=np.arange(-0.5,0.55,0.05),extend="both")
ax3.set_title("Potential Temperature-Conservative\nTemperature [°C] from GSW")
plt.show()
```

```python
model['gsw_o2sat']=xr.apply_ufunc(gsw.O2sol,models.gsw_as, 
                                  modelt.gsw_ct, 
                                  model.lev, 
                                  models.longitude, 
                                  models.latitude, 
                                  dask='parallelized',
                                  output_dtypes=[modelt.gsw_ct.dtype])*1.0245e-3
#model['gsw_o2sat']=xr.apply_ufunc(gsw.O2sol_SP_pt,models.so.isel(time=slice(-13,-1)).mean('time'), modelt.thetao.isel(time=slice(-13,-1)).mean('time'), dask='allowed')*1.0245e-3
# ValueError: applied function returned data with unexpected number of dimensions: 3 vs 4, for dimensions ('member_id', 'lev', 'j', 'i')

fig, (ax1,ax2,ax3) = plt.subplots(figsize=(15,4),ncols=3,nrows=1)
(model.isel(time=slice(-13,-1)).mean('time').isel(lev=0).o2sat*1e6).plot(ax=ax1,levels=np.arange(0,420,20),extend="both")
ax1.set_title("Oxygen Saturation [mmol/m3] from model")
(model.isel(lev=0).gsw_o2sat*1e3).plot(ax=ax2,levels=np.arange(0,420,20),extend="both")
ax2.set_title("Oxygen Saturation [mmol/m3] from GSW")
oanom=(model.o2sat*1e6-model.gsw_o2sat*1e3)
(oanom.isel(time=slice(-13,-1)).mean('time').isel(lev=0)).plot(ax=ax3,levels=np.arange(-5,5.5,0.5),extend="both")
ax3.set_title("Oxygen Saturation anomaly [model-gsw, mmol/m3] from GSW")
plt.show()
```
